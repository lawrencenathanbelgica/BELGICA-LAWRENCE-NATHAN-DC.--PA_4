```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_excel('board2.xlsx')
df
```

```python
sub = ['Electronics']

Inst = df.loc[(df['Hometown'] == 'Luzon') & (df['Track']  == 'Instrumentation') & (df['Electronics'] > 70), ['Name', 'GEAS', 'Electronics']]


Inst.to_csv('Inst.csv')
Inst
```

```python
# List of subjects to calculate the average
subs = ['Electronics', 'Math', 'GEAS', 'Communication']

df['Average'] = df[subs].mean(axis=1)

# Filter rows where Hometown is 'Mindanao', Gender is 'Female', and Average >= 55
Mindy = df.loc[(df['Hometown'] == 'Mindanao') & (df['Gender'] == 'Female') & (df['Average'] >= 55), ['Name', 'Track', 'Electronics', 'Average']]

Mindy.to_csv('Mindy.csv')
Mindy
```

```python
# Import matplotlib for plotting
import matplotlib.pyplot as plt
```

```python
# List of subjects to calculate the average
subs = ['Electronics', 'Math', 'GEAS', 'Communication']

#the mean score across the subjects
df['Average'] = df[subs].mean(axis=1)

# Set figure size (width=2, height=4 inches)
plt.figure(figsize=(2,4))
gender_avg = df.groupby('Gender')['Average'].mean()

# Create a bar chart of average scores by gender
plt.bar(gender_avg.index, gender_avg.values)

# Label the x-axis
plt.xlabel('Gender')
# Label the y-axis
plt.ylabel('Average Score')
```

```python
#Plot for Average by Hometown

subs = ['Electronics', 'Math', 'GEAS', 'Communication']
df['Average'] = df[subs].mean(axis=1)


# Set figure size (width=6, height=4 inches)
plt.figure(figsize=(6,4))
gender_avg = df.groupby('Hometown')['Average'].mean()

# Create a bar chart of average scores by hometown
plt.bar(gender_avg.index, gender_avg.values)

plt.xlabel('Hometown')
plt.ylabel('Average Score')
```

```python
# Plot for Average by Track

subs = ['Electronics', 'Math', 'GEAS', 'Communication']
df['Average'] = df[subs].mean(axis=1)

plt.figure(figsize=(6,4))
gender_avg = df.groupby('Track')['Average'].mean()

# Create a bar chart of average scores by track
plt.bar(gender_avg.index, gender_avg.values)

plt.xlabel('Track')
plt.ylabel('Average Score')
```
